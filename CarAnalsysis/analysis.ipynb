<h3> Remove outliers by normalizing Data </h3>
<p> On this step I will remove outliers and the usign the describe function check the price numbers look logical, From the prior describe function there was some vehicles sold at 3B+ </p>

```python
#normalize the data
z_scores = np.abs((df[['price', 'odometer']] - df[['price', 'odometer']].mean()) / df[['price', 'odometer']].std())

# Define a threshold for Z-score above which values are considered outliers
z_score_std_threshold = 2

# Create a boolean mask for outliers
outliers_mask = z_scores > z_score_std_threshold

# Identify rows with any outliers in 'price' or 'odometer'
rows_with_outliers = outliers_mask.any(axis=1)

# Remove rows with outliers
df_no_outliers = df[~rows_with_outliers]

#Describe to see results.
df_no_outliers.describe()
```

<h3> Limit Prices </h3>
<p> After removing outliers using Z-score there still cars with prices of 17M, as I assume this is not a specialized dealership selling classic or custom cars I think is better to remove all cars beyond 150K, This parameter needs to be adjusted based on the dealer needs, if it is an specialized dealership then you might want to remove the low value cars, also I am removing cars under 5K as a delearship might not be interested in selling this cars </p>

```python
cars=df_no_outliers[(df_no_outliers['price'] < 150000) & (df_no_outliers['price'] > 5000) & (df_no_outliers['year'] > 1995)]
cars.describe()
```

```python
import scipy.stats as stats
# Get unique paint colors in the filtered data
colors = filtered_cars['paint_color'].unique()

# Initialize lists to store non-empty group data
non_empty_groups = []
grouped_data = []

# Iterate through colors and filter data
for color in colors:
    group = filtered_cars[filtered_cars['paint_color'] == color]['price']
    if len(group) > 0:  # Check if the group has data points
        non_empty_groups.append(color)
        grouped_data.append(group)

# Perform ANOVA only on non-empty groups
if len(non_empty_groups) > 1:
    f_statistic, p_value = stats.f_oneway(*grouped_data)
    print("ANOVA F-Statistic:", f_statistic)
    print("ANOVA p-value:", p_value)
else:
    print("Insufficient data for ANOVA analysis")
```

```python
# Create a preprocessing pipeline
numeric_transformer = StandardScaler()

categorical_transformer = Pipeline(steps=[
     ('imputer', SimpleImputer(strategy='most_frequent')),  # Impute missing values with most frequent category
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# Define a range of polynomial degrees to evaluate
degrees = [1,2,3]
```

```python
# Create a pipeline for Linear Regression and Lasso
models = [
    ('linear', LinearRegression()),
   # ('lasso',  Lasso())
]

# Initialize a dictionary to store results
results = {'Model': [], 'Degree': [], 'Train MSE': [], 'Test MSE': [], 'Train R2': [], 'Test R2': [], 'Coeficients': [], 'Intercept': []}
# Iterate through models and degrees
for model_name, model in models:
    for degree in degrees:
        pipeline = Pipeline(steps=[
            ('preprocessor',  ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('poly', PolynomialFeatures(degree=degree, include_bias=False, interaction_only=True), numeric_features),
    ('cat', categorical_transformer, categorical_features)
])),
           # ('polynomial', PolynomialFeatures(degree=degree, include_bias=False, interaction_only=True)),
            ('model', model)
        ])
        
        # Fit the pipeline
        pipeline.fit(X_train, y_train)
        fitted_model = pipeline.named_steps['model']
        # Predict on train and test sets
        y_train_pred = pipeline.predict(X_train)
        y_test_pred = pipeline.predict(X_test)
        
        # Calculate MSE
        train_mse = mean_squared_error(y_train, y_train_pred)
        test_mse = mean_squared_error(y_test, y_test_pred)
        test_r2 = r2_score(y_train, y_train_pred)
        train_r2 =r2_score(y_train, y_train_pred)
        
        coefficients = fitted_model.coef_
        intercept = fitted_model.intercept_
        
        rounded_coefficients = [round(coef, 2) for coef in coefficients]
        rounded_intercept = round(intercept, 2)
        
        # Store results
        results['Model'].append(model_name)
        results['Degree'].append(degree)
        results['Train MSE'].append(train_mse)
        results['Test MSE'].append(test_mse)
        results['Train R2'].append(train_r2)
        results['Test R2'].append(test_r2)
        results['Coeficients'].append( rounded_coefficients )
        results['Intercept'].append( rounded_intercept  )

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Display the results
pd.set_option('display.max_columns', None)  # Display all columns
pd.set_option('display.max_rows', None)     # Display all rows
pd.set_option('display.max_colwidth', None) # Display full contents of columns


results_df
```
